## 2.2.2 성능 지표 선택
회긔 문제 -> **평균 제곱근 오차**<i>root mean square error<i/>(RMSE)
오차가 커질 수록 값은 더욱 커진다

$$ \text{RMSE}(\textbf{X}, h) = \sqrt{\displaystyle\frac{1}{m} \displaystyle\sum_{i=1}^m(h(x^{(i)}) - y^{(i)})^2}$$

- $m$: 데이터셋의 샘플 수
- $ x^{(i)} $ : 데이터 셋에 있는 $i$ 번째 샘플(레이블은 제외한)의 **전체 특성값의 벡터**
- $h$: 에측 함수



#### Norm(노름)
- $l_{2}$ 노름(Norm): 유클리디안 노름
- $l_{1}$ 노름(Norm): 절대값 노름

### 2.3.4 테스트 세트 만들기
테스트 세트로 일반화 오차를 추정하게 되면 **데이터 스누핑**(data snooping) 편향 발생  
-> 데이터를 무작위로 선택하여 선별 

```python
import numpy as np
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices, train_indices = shuffled_indices[:test_set_size], shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
len(train_set)
```

#### 무작위 샘플링
- 위와 같은 방식은 순수한 무작위 샘플링 방식  
- 데이터셋의 크기가 충분히 크다면(특히 특성수에 비해) 일반적으로 괜찮다.   
- 하지만, 적다면 **편향**이 생길 가능성이 높다 

  
#### 계층적 샘플링
- 샘플을 뽑을때도 그냥 뽑는 것이 아니라 데이터의 특성이 대표 되도록 뽑아야 한다.  
- 1000을 대상으로 설문을 할때에도 그냥 1000명이 아니라, 남성 vs. 여성 비율을 맞추어 뽑는다.  
- 이를 **계층적 샘플링** 이라고 한다.  
- 전체 인구는 **계층(strata)**이라는 동질의 그룹으로 나눈다. 
      


- `medium income`을 통해, 주택 가격을 에측하고 싶다.  
    -> 테스트 데이터 세트가 전체 데이터 세트의 있는 여러 income 카테고리를 잘 대표 되도록 샘플링 되어야 한다. 
- medium income은 **연속형** 숫자 이다.  
    -> 카테고리 attribute로 변환이 필요하다.(계층화를 위해)

- 각 계층을 잘 나눌 수 있는 카테고리가 필요하다.  
    -> 그렇치 않으면, **계층의 중요도를 측정하는데 편향**이 발생한다.  
    -> 너무 많은 계층으로 나누면 안된다. 각 계층이 충분히 커야 한다.
    
    

```python
# pandas 사용
housing['income_cat'] = pd.cut(housing['median_income'], bins=[0, 1.5, 3.0, 4.5, 6, np.inf], labels=[1,2,3,4,5])
```

```python
housing['income_cat'].hist()
plt.show()
```

```python
# sklearn 사용
# Stratified Shuffle Split 사용

from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(housing, housing['income_cat']):
    start_train_set = housing.loc[train_index]
    start_test_set = housing.loc[test_index]
```

```python
start_train_set['income_cat'].value_counts() / len(start_test_set)
```

```python
for set_ in (start_test_set, start_train_set):
    set_.drop('income_cat', axis=1, inplace=True)
```

### 2.5.1 데이터 정제
 - 대부분의 머신러닝 알고리즘은 누락된 특성을 다루지 못함  
 `total_bedroom` 특성 값이 없는 경우 이를 고쳐야함
  
 1. 해당 구역을 제거 -> `dropna`
 2. 전체 특성을 제거 -> `drop`
 3. 어떤 값으로 채운다(0, median, mean) -> `fillna`


- **dropna**
[pandas.DataFrame.dropna — pandas 1.2.0 documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html)

```python
housing.info()
```

#### 1) 전통적인 방법

```python
# 옵션 1
housing.dropna(subset = ['total_bedrooms'], how='any')
# subset 부분 집합에 포함된 속성일 경우 dropna 적용
# how: any(한개라도 N/A일경우), all(전부 N/A일 경우), axis=0를 기준 default

# 옵션 2
housing.drop('total_bedrooms', axis=1)

# 옵션 3
median = housing['total_bedrooms'].median()
housing['total_bedrooms'].fillna(median, inplace=True)
```

#### 2) sklearn 사용  
scikit-learn의 `Imputer`는 누락된 값을 손쉽게 다룰 수 있도록 해준다. `Imputer`를 이용해 Null 값에 중앙값을 채워보도록 하자.

**중앙값(median)** 은 numeric한 데이터 특성에서만 계산되므로 텍스트 특성인 `ocean_proximity`는 제거해준뒤 `Imputer.fit()`을 통해 null 값에 중앙값을 채워준다.


```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')

# 수치형만 저장
housing_num = housing.drop('ocean_proximity', axis=1) 
imputer.fit(housing_num)
```

`Imputer`는 각 특성의 중앙값(median)을 계산한 뒤 그 결과를 `.statistics_`에 저장한다.

```python
print(imputer.statistics_, housing_num.median().values, sep='\n')
```

`Imputer`의 `.transform()`을 이용해 Training Set에서 누락된 값을 중앙값으로 바꿀 수 있다. `.transform()`의 반환값은 `NumPy`의 `array`이다.

```python
X = imputer.transform(housing_num)
housing_tr = pd.DataFrame(X, columns=housing_num.columns, index=housing_num.index)
```

```python
housing_tr.info()
```

#### 일관성
: 모든 객체가 일관되고 단순한 인터페이스 공유

 - 추정기<sub>estimator</sub> :  `fit` 함수에 의해 수행되고 하나의 매개 변수로 하나의 데이터셋 전달, 인스턴스 변수로 저장됨
 - 변환기<sub>transformer</sub> : 데이터셋 변환 `transform`에 의해 사용됨. fit과 transform을 동시 사용가능한 `fit_transfrom()`도 존재
 - 예측기<sub>predictor</sub> : 데이터를 예측 할 떄 사용 `predict()`에 의해 사용  

#### 검사 기능
 - 추정기를 통해 사용가능한 하이퍼 파라미터는 Public 인스턴스로 사용가능.
 - 모든 추정기의 학습된 모델은 접미사를 붙혀 사용 가능 (예: `imputer.statistics_`)
